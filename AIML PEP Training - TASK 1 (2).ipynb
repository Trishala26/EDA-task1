# RANDOM UNDERSAMPLING

Undersampling is a technique used in machine learning to balance imbalanced datasets, where one class has significantly more samples than the other. In this case, the code is reducing the number of samples in the majority class to match the number of samples in the minority class.

```python
from imblearn.under_sampling import RandomUnderSampler
r=RandomUnderSampler(sampling_strategy=1)
x_res,y_res=r.fit_resample(x,y)

y_res.value_counts().plot.pie(autopct="%.2f").set_title("UNDERSAMPLING")
```

```python
y_res.value_counts()
```

# RANDOM OVERSAMPLING

Oversampling is a technique used in machine learning to balance imbalanced datasets, where one class has significantly fewer samples than the other. In this case, the code is increasing the number of samples in the minority class to match the number of samples in the majority class.

```python
from imblearn.over_sampling import RandomOverSampler
r=RandomOverSampler(sampling_strategy=1)
x_r,y_r=r.fit_resample(f,g)

y_r.value_counts().plot.pie(autopct="%.2f").set_title("OVERERSAMPLING")
```

```python
New_DataFrame=pd.merge(left=df1,right=dfs,how='inner',on='SK_ID_CURR',suffixes='_x')
```

```python
LIST_P = ["NAME_CONTRACT_TYPE_x", "CODE_GENDER", "FLAG_OWN_CAR", "FLAG_OWN_REALTY","NAME_TYPE_SUITE", "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", "NAME_FAMILY_STATUS", "NAME_HOUSING_TYPE", "NAME_CLIENT_TYPE","NAME_YIELD_GROUP", "CNT_CHILDREN"]
```

```python
def CAT_CAT_ANALYSIS_Prev(LIST_P, rot = False):
    for i in range(1, len(LIST_P)):
        res = pd.pivot_table(data = New_DataFrame, index = LIST_P[i], columns= 'NAME_CONTRACT_STATUS', values = "TARGET")
        plt.figure(figsize=[20,100])
        plt.subplot(len(LIST_P), 1, i)
        plt.subplots_adjust(hspace= 0.5)
        sns.heatmap(data = res, annot = True, cmap = "coolwarm")
        if rot:
            plt.xticks(rotation = 90)
        plt.show()
```
